```python
# Importación de bibliotecas
import pandas as pd

# Cargar el archivo proporcionado
from google.colab import files
uploaded = files.upload()

# Leer el archivo CSV
file_path = list(uploaded.keys())[0]
data = pd.read_csv(file_path)

# a. Extraer la información del archivo
print("Primeras filas del archivo:")
print(data.head())

# b. Mostrar la estructura y tipos de datos de cada columna
print("\nEstructura y tipos de datos:")
print(data.info())
print("\nConvertir la columna de fechas a datetime:")
data['date'] = pd.to_datetime(data['date'])
print(data.info())

# c. Determinar la cantidad de vacunas aplicadas de cada compañía
vaccines_data = data.groupby('vaccines')['total_vaccinations'].sum().reset_index()
print("\nCantidad de vacunas aplicadas por compañía:")
print(vaccines_data)

# d. Obtener la cantidad de vacunas aplicadas en todo el mundo
total_vaccines_world = data['total_vaccinations'].sum()
print("\nCantidad total de vacunas aplicadas en el mundo:")
print(total_vaccines_world)

# e. Calcular el promedio de vacunas aplicadas por país
average_vaccines_country = data.groupby('country')['total_vaccinations'].mean().reset_index()
print("\nPromedio de vacunas aplicadas por país:")
print(average_vaccines_country)

# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo
vaccines_29jan21 = data[data['date'] == '2021-01-29']['total_vaccinations'].sum()
print("\nCantidad de vacunas aplicadas el 29/01/21:")
print(vaccines_29jan21)

# g. Crear un dataframe con diferencias entre daily_vaccinations y daily_vaccinations_raw
data['diferencias'] = data['daily_vaccinations'] - data['daily_vaccinations_raw']
conDiferencias = data.copy()
print("\nDataframe con diferencias:")
print(conDiferencias.head())

# h. Obtener el periodo de tiempo entre la fecha más reciente y la más antigua
time_period = data['date'].max() - data['date'].min()
print("\nPeriodo de tiempo entre fechas más reciente y antigua:")
print(time_period)

# i. Crear un dataframe con la cantidad de vacunas utilizadas cada día
data['canVac'] = data['vaccines'].str.split(',').apply(len)
conCantidad = data.copy()
print("\nDataframe con cantidad de vacunas utilizadas cada día:")
print(conCantidad.head())

# j. Generar un dataframe con registros antes del 20 de diciembre de 2020
antes20 = data[data['date'] < '2020-12-20']
print("\nDataframe con registros antes del 20/12/2020:")
print(antes20.head())

# k. Obtener un dataframe con registros donde se haya utilizado la vacuna Pfizer
pfizer = data[data['vaccines'].str.contains('Pfizer')]
print("\nDataframe con registros donde se utilizó Pfizer:")
print(pfizer.head())

# l. Almacenar los dataframes generados en un archivo Excel
output_excel = 'resultadosReto.xlsx'
with pd.ExcelWriter(output_excel) as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)

# Descargar el archivo generado
files.download(output_excel)
```
